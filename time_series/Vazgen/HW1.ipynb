1) ** How do adaptive properties of SES change when smoothing parametr $\alpha$ converges to 1?**

Since $\hat{y}_{t+1}=\alpha y_t+(1-\alpha)\hat{y}_t$, thus if $\alpha$ converges to 1, then $\hat{y}_{t+1}\approx y_t$. So the prediction graph just repeat original with some shift.

2) ** What will be the optimal value of smoothing parameter of trend $\beta$ in Hold model for 1) time series with slowly  changing trend; 2) stochastic changing trend (look <a href="http://stochastictrend.blogspot.am/">here </a>)? **

The Holt model is the following
$$
\hat{y}_{t+d}=l_t+b_td,
$$
$$
l_t=\alpha y_t+(1-\alpha)(l_{t-1}+b_{t-1})
$$
$$
b_t=\beta(l_t-l_{t-1})+(1-\beta)b_{t-1}
$$
If time series has slowly changing trend, then we should take $b_t\approx b_{t-1}$, which means that $\beta$ should be near to $0$.

If time series has stochastic changing trend, it menas that trend changes significantly during the time. Therefore in this case the optimal value of beta will be near to $1$.

3) **Proof the equivalency of correction error form and smoothing form for unobserved component of TS in  Thei-Wage model (period of seasonality equals $p$):**

  $$1) l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} )=l_{t-1} + b_{t-1} + \alpha e_t;$$
  $$2) b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} = b_{t-1} + \alpha\beta e_t;$$
  $$3) s_t = \gamma (y_t-l_t) + (1-\gamma) s_{t-p} = s_{t-p} + \gamma(1-\alpha)e_t.$$

Proof:

Since $e_t=y_t-\hat{y}_t=y_t - s_{t-p}-l_{t-1} - b_{t-1}$, thus
$$
1)  l_t = \alpha (y_t - s_{t-p}) + (1-\alpha) (l_{t-1} + b_{t-1} )=l_{t-1} + b_{t-1}+\alpha (y_t - s_{t-p})-\alpha (l_{t-1} + b_{t-1} )=l_{t-1} + b_{t-1}+\alpha (y_t - s_{t-p}-l_{t-1} - b_{t-1})=l_{t-1} + b_{t-1} + \alpha e_t.
$$

From the previous point follows that
$$
2) b_t = \beta (l_{t} - l_{t-1} ) + (1-\beta) b_{t-1} = \beta (l_{t} - l_{t-1} ) + b_{t-1}-\beta b_{t-1}=b_{t-1}+\beta(l_{t} - l_{t-1}-b_{t-1})=b_{t-1} + \alpha\beta e_t.
$$
From the first point follows that
$$
3) s_t=\gamma (y_t-l_t) + (1-\gamma) s_{t-p}=\gamma (y_t-l_t)+s_{t-p}-\gamma s_{t-p}=s_{t-p}+\gamma(y_t-l_t-s_{t-p})=s_{t-p}+\gamma(y_t-s_{t-p}-l_{t-1}-b_{t-1}-\alpha e_t)=s_{t-p}+\gamma(e_t-\alpha e_t)=s_{t-p} + \gamma(1-\alpha)e_t.
$$
4) **Find the correction error form for unobserved component of TS in ES with additive damped trend:**

$$\hat{y}_{t+d|t} = l_t + \left(\phi + \phi^2 + \dots + \phi^{d}\right) b_t.$$

**Unobserved components are (in smoothing form): **

$$ l_{t}       = \alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right),$$
$$ b_t         = \beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}.$$

In this model we have $\hat{y}_{t|t-1}=l_{t-1}+\phi b_{t-1}$, thus
$$
e_t=y_{t}-\hat{y}_{t|t-1}=y_{t}-l_{t-1}-\phi b_{t-1}.
$$
It follows that
$$
l_t=\alpha y_t + \left(1-\alpha\right) \left(l_{t-1} +\phi b_{t-1}\right)=\alpha y_t+l_{t-1} +\phi b_{t-1}-\alpha \left(l_{t-1} +\phi b_{t-1}\right)=l_{t-1} +\phi b_{t-1}+\alpha\left(y_{t}-l_{t-1}-\phi b_{t-1}\right)=l_{t-1} +\phi b_{t-1}+\alpha e_t.
$$
From this we obtain $l_t-l_{t-1}=\phi b_{t-1}+\alpha e_t$, therefore
$$
b_t=\beta \left(l_t - l_{t-1}\right) + \left(1-\beta\right)\phi b_{t-1}=\beta \left(\phi b_{t-1}+\alpha e_t\right) + \left(1-\beta\right)\phi b_{t-1}=\phi b_{t-1}+\alpha\beta e_t.
$$
