1) **Find equation for $y_t$ that corresponds to the following ARIMA model**:
 - 1) ARIMA $(p=0,d=k,q=1)$,
 
 
 ARIMA(p,d,q) model is the following
$$
\phi\left(B\right)\nabla^d y_t=\theta(B)y_t, 
$$
where $By_t=y_{t-1}$ and
$$
\phi\left(B\right)=1-\phi_1B-\phi_2B^2-\ldots-\phi_pB^p,
$$
$$
\theta\left(B\right)=1+\theta_1B+\theta_2B^2+\ldots+\theta_qB^q.
$$
So if $p=0,\ d=k,\ q=1$, then
$$
\nabla^k y_t=\left(1+\theta_1B\right)\varepsilon_t, 
$$
which is the same that
$$
\left(1-B\right)^k y_t=\varepsilon_t+\theta_1\varepsilon_{t-1}
$$
$$
\sum\limits_{n=0}^k\left(-1\right)^nB^ny_t=\varepsilon_t+\theta_1\varepsilon_{t-1}
$$
$$
\sum\limits_{n=0}^k\left(-1\right)^ny_{t-n}=\varepsilon_t+\theta_1\varepsilon_{t-1}
$$
$$
y_t=\sum\limits_{n=1}^k\left(-1\right)^{n+1}y_{t-n}+\varepsilon_t+\theta_1\varepsilon_{t-1}.
$$
 - 2) ARIMA $(p=0,d=k+1,q=1)(P=1, D=1, Q=0)_s$,

ARIMA $(p,d,q)(P,D,Q)_s$ model is the following
$$
\Phi_p(B^s)\phi(B)\nabla_s^D\nabla^{k+1}\left(y_t-\mu\right)=\Theta_Q\left(B^s\right)\theta(B)\varepsilon_t
$$
where $B$, $\phi$, $\theta$ are mentioned above and
$$
\Phi_P\left(B^s\right)=1-\Phi_1B^s-\Phi_2 B^s-\ldots-\Phi_PB^{PS}
$$
$$
\Theta_P\left(B^s\right)=1+\Theta_1B^s+\Theta_2 B^s+\ldots+\Theta_QB^{QS}
$$
So if $p=0, d=k+1, q=1$ and $P=1, D=1, Q=0$, then we will have
$$
\left(1-\Phi_1B^s\right)\nabla_s^1\nabla^{k+1}\left(y_t-\mu\right)=\left(1+\theta_1B\right)\varepsilon_t.
$$
Note that
$$
\left(1-\Phi_1B^s\right)\nabla_s^1\nabla^{k+1}\left(y_t-\mu\right)=\left(1-\Phi_1B^s\right)\left(1-B\right)\left(1-B^s\right)\left(y_t-\mu\right)=\left(1-\Phi_1B^s\right)\left(1-B\right)\left(y_t-y_{t-s}\right)=
\left(1-\Phi_1B^s\right)\left(y_t-y_{t-1}-y_{t-s}+y_{t-s-1}\right)
$$
$$
=y_t-y_{t-1}-y_{t-s}+y_{t-s-1}-\Phi_1\left(y_{t-s}-y_{t-s-1}-y_{t-2s}+y_{t-2s-1}\right)=y_t-y_{t-1}-\left(1+\Phi_1\right)y_{t-s}+\left(1+\Phi_1\right)y_{t-s-1}+\Phi_1y_{t-2s}-\Phi_1y_{t-2s-1}.
$$
So in this case our model is
$$
y_t=y_{t-1}+\left(1+\Phi_1\right)y_{t-s}-\left(1+\Phi_1\right)y_{t-s-1}-\Phi_1y_{t-2s}+\Phi_1y_{t-2s-1}+\varepsilon_t+\theta_1\varepsilon_{t-1}.
$$
